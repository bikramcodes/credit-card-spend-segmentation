```python
# Loading the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df_clusters = pd.read_csv('../data/processed/cc_clusters.csv')
```

```python
features = ['BALANCE', 'PURCHASES', 'PURCHASES_TRX', 'PAYMENTS', 'CREDIT_LIMIT']

for col in features:
    plt.figure(figsize=(8, 4))
    sns.barplot(x='Cluster', y=col, data=df_clusters, palette='Set2')
    plt.title(f'Average {col} per Cluster')
    plt.ylabel(col)
    plt.xlabel('Cluster')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
