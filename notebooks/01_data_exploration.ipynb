```python
# Loading the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style='whitegrid')
%matplotlib inline
```

```python
# Loading Data

df = pd.read_csv("../data/raw/CC GENERAL.csv")
```

```python
# Checking the shape of the data set

print("Shape:", df.shape)
```

```python
# Describing the data

print("\nSummary Stats:\n\n", df.describe().T)
```

```python
# Checking the missing values

print("\nMissing values:\n\n", df.isnull().sum())
```

```python
# Checking the data information 

print("\nData Information:\n")
df.info()
```

```python
# Visualize missing data

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Sample records
df.head()
```

```python
# Check for duplicates

print(df.duplicated().sum())
```

```python
# Visualizing key numerical variables

cols_to_plot = ['BALANCE', 'PURCHASES', 'CREDIT_LIMIT', 'PAYMENTS']

df[cols_to_plot].hist(bins=30, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Key Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap

plt.figure(figsize=(14, 10))
corr = df.select_dtypes(exclude='object').corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```
